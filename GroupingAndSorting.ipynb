```python
import pandas as pd
sales = pd.read_csv("Data/sales.csv")
sales
```

# However, often we want to group our data, and then do something specific to the group the data is in.

# As you'll learn, we do this with the groupby() operation. We'll also cover some additional topics, such as more complex ways to index your DataFrames, along with how to sort your data.

```python
sales.groupby('Sales').Sales.count() #long form of value_counts()
#this code will give us unique values as the index and occureacne as the values
```

```python
sales.groupby('Sales').Sales.min()
```

```python
sales.groupby('Sales').Sales.size()
```

```python
sales.groupby('Sales').Marketing.min() #for each index, give us minimun
```

sales.groupby('Marketing').apply(lambda df: df.Sales.iloc[0])

```python
sales.groupby(['Marketing', 'Sales']).apply(lambda df: df.loc[df.Profit.idxmax()])
#pick out best profit by marketing and sales
```

# Another groupby() method worth mentioning is agg(), which lets you run a bunch of different functions on your DataFrame simultaneously. For example, we can generate a simple statistical summary of the dataset as follows:

```python
sales.groupby(['State']).Profit.agg([len, min, max]) #really nice
```

# multiple indexes

```python
sales.groupby(['State','Product']).Profit.agg([len, min, max])
```

```python
sales.groupby(['State','Product']).Profit.agg([len, min, max]).reset_index()
```

# note that when something is groupby, the index is reorderd by index and not value, we can change this by using sort_values()

```python
sales.groupby(['State','Product']).Profit.agg([len, min, max]).reset_index().sort_values(by="len")
#sort by len in the previous cell
```

```python
sales.groupby(['State','Product']).Profit.agg([len, min, max]).reset_index().sort_index()#to sort by index 
```

```python
#sort by more than one colume
sales.groupby(['State','Product']).Profit.agg([len, min, max]).reset_index().sort_values(by=['len','min'])
```
